```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
```

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='latin1')
df.head(20)
```

```python
remove_columns = ['approxdate', 'resolution', 'related']

df = df.drop(df.loc[:,remove_columns], axis = 1)
df.head(10)
```

```python
df.describe()
```

```python
df.dtypes
```

```python

missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
print("Missing values in each column:")
print(missing_values)
print("\nPercentage of missing values in each column:")
print(missing_percentage)

# Assess critical columns
interested_columns = ['iyear', 'imonth', 'iday', 'country', 'region', 'extended', 'nreleased', 'INT_IDEO', 'INT_LOG', 'INT_MISC']
missing_values_interested = df[interested_columns].isnull().sum()
print("\nMissing values in interested columns:")
print(missing_values_interested)

# Strategy for handling missing values
initial_count = len(df)

# Option 1: Drop rows with critical missing values
df.dropna(subset=['iyear', 'imonth', 'iday', 'country', 'region'], inplace=True)

# Check rows left
if len(df) < initial_count * 0.85:  # More than 15% dropped
    print("Too many rows removed, consider filling missing values instead.")
else:
    print(f"Dropped rows, current count: {len(df)}")

# Option 2: Fill missing values
df['nreleased'].fillna(df['nreleased'].median(), inplace=True)
df['country'].fillna(df['country'].mode()[0], inplace=True)
df['region'].fillna(df['region'].mode()[0], inplace=True)
df['extended'].fillna(0, inplace=True)  # Assuming 0 for neutral state

# Check the final dataframe
print("\nFinal DataFrame shape:")
print(df.shape)
print("\nMissing values after handling:")
print(df.isnull().sum())
```

```python
df['nreleased'].fillna(0, inplace=True)
```

```python
df[['INT_IDEO', 'INT_LOG', 'INT_MISC']].isnull().sum()
```

```python
X = df[['iyear', 'imonth', 'iday', 'country', 'region', 'extended', 'nreleased', 'INT_LOG', 'INT_MISC']]
y = df['INT_IDEO']
```

```python

categorical_cols = ['country', 'region']
ohe = OneHotEncoder(drop='first', sparse_output=False)
X_encoded = ohe.fit_transform(X[categorical_cols])
X_encoded_df = pd.DataFrame(X_encoded, columns=ohe.get_feature_names_out(categorical_cols))
X_final = pd.concat([X.drop(columns=categorical_cols).reset_index(drop=True), X_encoded_df], axis=1)
X_final.head(10)
```

```python
features_train, features_test, target_train, target_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
```

```python
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(features_train, target_train)
```

```python
predicted_lr = lr_model.predict(features_test)
```

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(features_train, target_train)
```

```python
predicted_rf = rf_model.predict(features_test)
```

```python
svc_model = SVC()
svc_model.fit(features_train, target_train)
```

```python
predicted_svc = svc_model.predict(features_test)
```

```python
recall_lr = recall_score(target_test, predicted_lr, average='weighted')
accuracy_lr = accuracy_score(target_test, predicted_lr)
f1_lr = f1_score(target_test, predicted_lr, average='weighted')
matrix_lr = confusion_matrix(target_test, predicted_lr)
report_lr = classification_report(target_test, predicted_lr)
```

```python
recall_rf = recall_score(target_test, predicted_rf, average='weighted')
accuracy_rf = accuracy_score(target_test, predicted_rf)
f1_rf = f1_score(target_test, predicted_rf, average='weighted')
matrix_rf = confusion_matrix(target_test, predicted_rf)
report_rf = classification_report(target_test, predicted_rf)
```

```python
recall_svc = recall_score(target_test, predicted_svc, average='weighted')
accuracy_svc= accuracy_score(target_test, predicted_svc)
f1_svc = f1_score(target_test, predicted_svc, average='weighted')
matrix_svc = confusion_matrix(target_test, predicted_svc)
report_svc = classification_report(target_test, predicted_svc)
```
